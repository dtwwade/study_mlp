```python
#draw the histagram of Gini coefficient in 2007, 2012 and 2014 or latest vs. country names
Gini_2010 = df.groupby("Country")['Gini_2007','Gini_2012','Gini_2014 or latest'].sum()
Gini_2010.plot(kind='bar', color=['r','b','y']) #set bar colors
plt.ylabel('Gini_coefficient') #Y-axis label
plt.title('Gini coefficient in 2007,2012 and 2014 or latest')
fig = plt.gcf()
fig.set_size_inches(20, 5) #adjust the size of the histagram
plt.show()
```

```python
#draw the histagram of poverty rate in 2007, 2012 and 2014 or latest vs. country names
Poverty_rate = df.groupby("Country")['Poverty rate_2007', 'Poverty rate_2012', 'Poverty rate_2014 or latest'].sum()
Poverty_rate.plot(kind='bar', color=['r','b','y','g']) #set bar colors
plt.ylabel('Poverty rate/%') #Y-axis label
plt.title('Poverty rate in 2007,2012 and 2014 or latest')
fig = plt.gcf()
fig.set_size_inches(20, 5)#adjust the size of the histagram
plt.show()
```

```python
#draw the histagram of poverty rate in 2014 or latest by age group vs. country names
Age_group = df.groupby("Country")['Children(< 18)', 'Youth(18-25)', 'Adult(26-65)', 'Elderly(> 65)','Working poor'].sum()
Age_group.plot(kind='bar', color=['r','b','y','g']) #set bar colors
plt.ylabel('Poverty rate/%') #Y-axis label
plt.title('Poverty rate (relative threshold),by age group, 2014 or latest')
fig = plt.gcf()
fig.set_size_inches(20, 5)#adjust the size of the histagram
plt.show()
```

```python
#The data set for making prediction

#XTrain = df[['2007-2010_Bottom 10%', '2007-2010_Top 10%', '2007-2010_Median income', '2007-2010_Mean income']].values
#yTrain = df['Gini_2010'].values

XTrain = df[['2010-2014_Bottom 10%', '2010-2014_Top 10%', '2010-2014_Median income', '2010-2014_Mean income',
            'Bottom10%', 'Bottom20%', 'Bottom40%', 'Top40%','Top20%','Top10%']].values 
yTrain = df['Gini_2014 or latest'].values
```

**Making prediction with linear regression**

```python
from sklearn import linear_model

# Learn from the data and predict yhat
lm = linear_model.LinearRegression()
lm.fit(XTrain, yTrain)
yhat = lm.predict(XTrain)

# Print the coefficients
print(np.round(lm.coef_,2),'\n')

sse = np.power(yTrain-yhat,2).sum()
tss = np.power(yTrain-yTrain.mean(),2).sum()

print('SSE: ')
print(np.round(sse, 3), '\n')

print('TSS: ')
print(np.round(tss, 3), '\n')

print('R^2: ')
print(np.round(1-sse/tss, 3), '\n')
```

**Making prediction with LASSO**

```python

from sklearn import linear_model as lm

# running predictions for 6 values of lambda/alpha
# Print the coefficients of model with different alpha
alphas = [10**-4, 10**-3, 10**-2,10**-1, 0.2,1]

for a in alphas:
    lr= lm.Lasso(alpha=a) #LASSO
    lr.fit(XTrain, yTrain)
    yhat = lr.predict(XTrain)
    
    sse = np.power(yTrain-yhat,2).sum()
    tss = np.power(yTrain-yTrain.mean(),2).sum()
    print('R^2: ')
    print(np.round(1-sse/tss, 3), '\n')
```

**Making prediction with randomforest regression**
