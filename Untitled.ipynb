```python
XTrain = df[['2007-2010_Bottom 10%', '2007-2010_Top 10%', '2007-2010_Median income', '2007-2010_Mean income']].values
yTrain = df['Gini_2010'].values
XTest = df[['2010-2014_Bottom 10%', '2010-2014_Top 10%', '2010-2014_Median income', '2010-2014_Mean income']].values
yTest = df['Gini_2014 or latest'].values
```

```python
from sklearn import linear_model

# Learn from the data and predict yhat
lm = linear_model.LinearRegression()
lm.fit(XTrain, yTrain)
yhat = lm.predict(XTest)

# Print the coefficients
print(np.round(lm.coef_,2),'\n')


print('')     

sse = np.power(yTest-yhat,2).sum()
tss = np.power(yTest-yTest.mean(),2).sum()

print('SSE: ')
print(np.round(sse, 3), '\n')

print('TSS: ')
print(np.round(tss, 3), '\n')

print('R^2: ')
print(np.round(1-sse/tss, 3), '\n')
```

```python
from sklearn import linear_model as lm

#Plot predictions for 6 values of lambda/alpha
alphas = [10**-5, 10**-4, 10**-3, 10**-2,10**-1, 1]

for a in alphas:
    lr= lm.Lasso(alpha=a)
    lr.fit(XTrain, yTrain)
    yhat = lr.predict(XTest)
    
    sse = np.power(yTest-yhat,2).sum()
    tss = np.power(yTest-yTest.mean(),2).sum()
    print('R^2: ')
    print(np.round(1-sse/tss, 3), '\n')
```
